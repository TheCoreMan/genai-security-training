# Lab 3: Interpretability Attacks

## Objectives
- Attack SHAP explanations
- Manipulate feature importance
- Test explanation robustness

```python
import torch
import torch.nn as nn
import numpy as np

# Detect device (supports CUDA, Apple Silicon MPS, and CPU)
if torch.cuda.is_available():
    device = 'cuda'
elif hasattr(torch.backends, 'mps') and torch.backends.mps.is_available():
    device = 'mps'
else:
    device = 'cpu'
```

## Part 1: Simple Explanation Attack

```python
model = nn.Sequential(nn.Linear(10, 2)).to(device)

def simple_explanation(model, x):
    """Gradient-based feature importance"""
    x.requires_grad = True
    output = model(x)
    output[0, 1].backward()
    return x.grad.abs()

x = torch.randn(1, 10, device=device)
importance = simple_explanation(model, x)
print(f'Feature importance: {importance}')
```
