```python
import torch
import torch.nn as nn

# Detect device (supports CUDA, Apple Silicon MPS, and CPU)
if torch.cuda.is_available():
    device = 'cuda'
elif hasattr(torch.backends, 'mps') and torch.backends.mps.is_available():
    device = 'mps'
else:
    device = 'cpu'
```

## Part 1: Cross-Modal Attack

```python
# Simulate multimodal model
class MultimodalModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.vision = nn.Linear(784, 128)
        self.text = nn.Linear(100, 128)
        self.fusion = nn.Linear(256, 10)
    
    def forward(self, img, text):
        v = self.vision(img.view(-1, 784))
        t = self.text(text)
        return self.fusion(torch.cat([v, t], dim=1))

model = MultimodalModel().to(device)
print('✓ Multimodal model created')
```
