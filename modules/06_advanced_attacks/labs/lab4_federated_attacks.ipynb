```python
import torch
import torch.nn as nn

# Detect device (supports CUDA, Apple Silicon MPS, and CPU)
if torch.cuda.is_available():
    device = 'cuda'
elif hasattr(torch.backends, 'mps') and torch.backends.mps.is_available():
    device = 'mps'
else:
    device = 'cpu'
```

## Part 1: Simulate FL

```python
# Simulate clients
n_clients = 10
models = [nn.Linear(10, 2).to(device) for _ in range(n_clients)]

# Aggregate gradients
def aggregate(models):
    avg_params = {}
    for name, param in models[0].named_parameters():
        avg_params[name] = torch.stack([m.state_dict()[name] for m in models]).mean(0)
    return avg_params

print(f'Simulating {n_clients} clients')
```
