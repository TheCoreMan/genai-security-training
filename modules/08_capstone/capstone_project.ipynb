```python
# Document system architecture
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Create architecture diagram (simple flow chart)
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_xlim(0, 10)
ax.set_ylim(0, 6)
ax.axis('off')

# Define components and positions
components = [
    {'name': 'User', 'pos': (1, 3)},
    {'name': 'API Gateway', 'pos': (3.5, 3)},
    {'name': 'LLM', 'pos': (6, 3)},
    {'name': 'Vector DB', 'pos': (6, 1)},
    {'name': 'Response', 'pos': (8.5, 3)}
]

# Draw boxes
for comp in components:
    box = mpatches.FancyBboxPatch(
        (comp['pos'][0] - 0.6, comp['pos'][1] - 0.3),
        1.2, 0.6,
        boxstyle='round,pad=0.1',
        facecolor='lightblue',
        edgecolor='black',
        linewidth=2
    )
    ax.add_patch(box)
    ax.text(comp['pos'][0], comp['pos'][1], comp['name'],
            ha='center', va='center', fontsize=10, weight='bold')

# Draw arrows
arrows = [
    (1.6, 3, 2.9, 3),      # User -> API Gateway
    (4.1, 3, 5.4, 3),      # API Gateway -> LLM
    (6, 2.7, 6, 1.6),      # LLM -> Vector DB
    (6.6, 3, 7.9, 3)       # LLM -> Response
]

for x1, y1, x2, y2 in arrows:
    ax.annotate('', xy=(x2, y2), xytext=(x1, y1),
                arrowprops=dict(arrowstyle='->', lw=2, color='black'))

plt.title('System Architecture', fontsize=14, weight='bold')
plt.tight_layout()
plt.show()

print("\n✓ Architecture diagram created!")
```

### 4.3 Risk Scoring

```python
# Calculate risk scores
def calculate_risk_score(likelihood, impact):
    """Calculate risk score (1-10)"""
    risk_matrix = {
        ('Low', 'Low'): 2,
        ('Low', 'Medium'): 3,
        ('Low', 'High'): 4,
        ('Medium', 'Low'): 4,
        ('Medium', 'Medium'): 6,
        ('Medium', 'High'): 8,
        ('High', 'Low'): 6,
        ('High', 'Medium'): 8,
        ('High', 'High'): 10
    }
    return risk_matrix.get((likelihood, impact), 5)

# Calculate for each vulnerability
for vuln in vulnerabilities:
    vuln['risk_score'] = calculate_risk_score('High', 'High')  # Update with actual values

print('Risk Scores:')
for vuln in vulnerabilities:
    print(f"  {vuln['id']}: {vuln['risk_score']}/10")
```
